```python
df1=df1.dropna(axis=0)
```

```python
#df1['CCR Efficiency']=pd.to_numeric(df1['CCR Efficiency'])
```

```python
Y_ccr=df1['CCR Efficiency']
Y_bcc=df1['BCC Efficinecy']
```

```python
X=df1.drop('CCR Efficiency',axis=1)
```

```python
X=X.drop('BCC Efficinecy',axis=1)
```

```python
X
```

```python
Y_bcc
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_bcc, test_size = 0.25, random_state = 42)
```

```python

# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
rfr = RandomForestRegressor(n_estimators = 800, random_state = 100)
  
# fit the regressor with x and y data
rfr.fit(X_train, Y_train)  
```

```python

y_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, rfr.predict(X_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

features = X.columns
# Obtain just the first tree
first_tree = rfr.estimators_[0]

plt.figure(figsize=(40,12))

tree.plot_tree(first_tree,
               feature_names=features,
               fontsize=8, 
               filled=True, 
               rounded=True);
plt.savefig("D:/tree_bank_ccr.jpg")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
print('Mean Absolute Error:', mean_absolute_error(Y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(Y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(Y_test, y_pred)))
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(Y_test, y_pred), 2))
```

```python

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, ensemble
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}
```

```python
reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(X_train, Y_train)

mse = mean_squared_error(Y_test, reg.predict(X_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

```python
feature_importance = reg.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + 0.5
fig = plt.figure(figsize=(30, 10))
plt.subplot(1, 2, 1)
plt.barh(pos, feature_importance[sorted_idx], align="center")
plt.yticks(pos, np.array(df1.columns)[sorted_idx])
plt.title("Feature Importance (MDI)")

result = permutation_importance(
    reg, X_test, Y_test, n_repeats=10, random_state=42, n_jobs=2
)
#sorted_idx = result.importances_mean.argsort()
#plt.subplot(1, 2, 2)
#plt.boxplot(
 #   result.importances[sorted_idx].T,
  #  vert=False,
   # labels=np.array(df1.columns)[sorted_idx],
#)
#plt.title("Permutation Importance")
#fig.tight_layout()
#plt.savefig("D:/XGBoost_feature_importance_ccr.jpg")
#plt.show()
```
